```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

emp_count = new_data.groupby('Region')['EmployeeID'].count().reset_index(name='Employee_Count')

sales_sum = new_data.groupby('Region')['Sales_Amount'].sum().reset_index(name='Total_Sales')


region_df = pd.merge(emp_count, sales_sum, on='Region')


sns.scatterplot(data=region_df, x='Employee_Count', y='Total_Sales')
plt.title('Region wise Sales vs Employee Count')
plt.xlabel('Number of Employees')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python

dept_sales = new_data.groupby('Department')['Sales_Amount'].sum().reset_index()

sns.barplot(data=dept_sales, x='Department', y='Sales_Amount')
plt.xticks(rotation=45)
plt.title('Department wise Sales Generated')
plt.show()
```

```python
years_sales = new_data.groupby('YearsAtCompany')['Sales_Amount'].mean().reset_index()

sns.lineplot(data=years_sales, x='YearsAtCompany', y='Sales_Amount')
plt.title('Years at Company vs Avg Sales')
plt.show()
```

```python
ot_sales = new_data.groupby('OverTime')['Sales_Amount'].mean().reset_index()

sns.barplot(data=ot_sales, x='OverTime', y='Sales_Amount')
plt.title('Overtime vs Average Sales')
plt.show()
```

```python
attr_sales = new_data.groupby('Attrition')['Sales_Amount'].mean().reset_index()

sns.barplot(data=attr_sales, x='Attrition', y='Sales_Amount')
plt.title('Attrition vs Average Sales')
plt.show()
```
